## 30. 形態素解析結果の読み込み

形態素解析結果(neko.txt.mecab)を読み込むプログラムを実装せよ。ただし、各形態素は表層形（surface）、基本形（base）、品詞（pos）、品詞細分類（pos1）をキーとするマッピング型に格納し、1文を形態素（マッピング型）のリストとして表現せよ。第4章の残りの問題では、ここで作ったプログラムを活用せよ。

```python
sentences = []
sentence = []
with open("neko.txt.mecab", "r") as f:
    for line in f:
        if line == "EOS\n":
            if len(sentence)>0:
                sentences.append(sentence)
            sentence = []
        else:
            surface, features = line.split("\t")
            features = features.split(",")
            morpheme_dict = {"surface": surface, "base": features[6], "pos": features[0], "pos1": features[1]}
            sentence.append(morpheme_dict)
```

## 31. 動詞

動詞の表層形をすべて抽出せよ。

```python
verb_list = []
for sentence in sentences:
    for morpheme in sentence:
        if morpheme['pos'] == '動詞':
            verb_list.append(morpheme['surface'])

for surface in verb_list:
    print(surface)
```

## 32. 動詞の原形

動詞の原形をすべて抽出せよ。

```python
base_list = []
for sentence in sentences:
    for morpheme in sentence:
        if morpheme['pos'] == "動詞":
            base_list.append(morpheme['base'])

for base in base_list:
    print(base)
```

## 33. サ変名詞

サ変接続の名詞をすべて抽出せよ。

```python
sahen_list = []
for sentence in sentences:
    for morpheme in sentence:
        if morpheme['pos1'] == "サ変接続":
            sahen_list.append(morpheme['surface'])

for surface in sahen_list:
    print(surface)
```
